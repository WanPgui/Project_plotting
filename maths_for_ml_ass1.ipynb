```python
# Load the dataset
data = pd.read_csv('fuel_econ.csv')

# Display the first few rows of the dataset
data.head(6)
```

```python
# Plot a histogram for 'comb' (combined fuel efficiency)
plt.figure(figsize=(10, 6))
sns.histplot(data['comb'], bins=30, kde=True)
plt.title('Combined Fuel Efficiency Distribution')
plt.xlabel('Combined Fuel Efficiency (MPG)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot a heatmap to show correlations between numerical columns
plt.figure(figsize=(12, 8))
corr = data.corr(numeric_only=True)  # Silencing the FutureWarning for numeric-only columns
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Vehicle Features')
plt.show()
```

```python
# Save the histogram as a PNG file
plt.figure(figsize=(10, 6))
sns.histplot(data['comb'], bins=30, kde=True)
plt.title('Combined Fuel Efficiency Distribution')
plt.xlabel('Combined Fuel Efficiency (MPG)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('combined_fuel_efficiency_histogram.png')
```

```python
# Save the heatmap as a PNG file
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Vehicle Features')
plt.savefig('correlation_heatmap.png')
```
